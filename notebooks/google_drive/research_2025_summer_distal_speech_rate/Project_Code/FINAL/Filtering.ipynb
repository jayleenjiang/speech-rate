```python
# === PATH to EEGData Folder (with annotated .fif files) ===
root_dir = '/content/drive/MyDrive/2025_Summer (Distal Speech Rate)/Project_Code/EEGData'

# === Loop through all participant folders ===
participants = sorted([p for p in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, p))])

for pid in participants:
    input_fif = os.path.join(root_dir, pid, f"{pid}_annotated_raw.fif")
    output_fif = os.path.join(root_dir, pid, f"{pid}_filtered_raw.fif")

    if not os.path.exists(input_fif):
        print(f"[!] {pid}: annotated file not found.")
        continue

    print(f"\n=== Filtering {pid} ===")

    try:
        raw = mne.io.read_raw_fif(input_fif, preload=True)

        # Remove 60 Hz line noise
        # raw.notch_filter(freqs=60.0)

        # High-pass filter at 0.1 Hz, Low-pass filter at 50 Hz
        raw.filter(l_freq=0.1, h_freq=50.0)

        # Save filtered file
        raw.save(output_fif, overwrite=True)
        print(f"✓ Saved filtered data to {output_fif}")
    except Exception as e:
        print(f"[ERROR in {pid}]: {e}")
```
