```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bootstrap
import mne

# === Configurations ===
root_dir = '/content/drive/MyDrive/2025_Summer (Distal Speech Rate)/Project_Code/EEGData'
participants = sorted([p for p in os.listdir(root_dir) if p.isdigit() and 10 <= int(p) <= 27])
conditions = ['Wd2N', 'Wd2E']  # Using case-insensitive match
epo_suffix = '_O1_5sd_epo.fif'
n_resamples = 1000
ci_level = 0.95

# === Scalp regions ===
regions = {
    'Left Anterior':     ['E48', 'E43', 'E44', 'E38', 'E32', 'E128', 'E25', 'E22', 'E26', 'E23', 'E27', 'E33', 'E34', 'E35', 'E40', 'E39', 'E28', 'E24', 'E20'],
    'Medial Anterior':   ['E12', 'E5', 'E19', 'E11', 'E4', 'E18', 'E16', 'E10', 'E15', 'E21', 'E14', 'E17', 'E127', 'E126'],
    'Right Anterior':    ['E117', 'E118', 'E124', 'E110', 'E109', 'E116', 'E123', 'E3', 'E9', 'E2', 'E122', 'E115', 'E114', 'E121', 'E1', 'E8', 'E125', 'E120', 'E113', 'E119'],
    'Left Central':      ['E29', 'E30', 'E41', 'E45', 'E46', 'E42', 'E47', 'E49', 'E37', 'E36'],
    'Medial Central':    ['E6', 'E7', 'E13', 'E31', 'E106', 'E112', 'E80', 'E54', 'E55', 'E79'],
    'Right Central':     ['E87', 'E93', 'E98', 'E102', 'E103', 'E107', 'E108', 'E104', 'E105', 'E111'],
    'Left Posterior':    ['E50', 'E51', 'E52', 'E53', 'E56', 'E57', 'E58', 'E59', 'E60', 'E63', 'E64', 'E65', 'E66', 'E68', 'E69', 'E70', 'E73'],
    'Medial Posterior':  ['E61', 'E62', 'E67', 'E72', 'E77', 'E71', 'E76', 'E75', 'E74', 'E78', 'E82', 'E81'],
    'Right Posterior':   ['E83', 'E84', 'E85', 'E86', 'E88', 'E89', 'E90', 'E91', 'E92', 'E94', 'E95', 'E96', 'E97', 'E99', 'E100', 'E101']
}
region_order = list(regions.keys())

# === Collect ERP data across participants ===
region_data = {cond: {region: [] for region in region_order} for cond in conditions}
times = None

for pid in participants:
    fpath = os.path.join(root_dir, pid, f"{pid}{epo_suffix}")
    if not os.path.exists(fpath):
        print(f"[{pid}] ❌ Missing file.")
        continue

    try:
        epochs = mne.read_epochs(fpath, preload=True)
        epochs.set_montage('GSN-HydroCel-128')

        # Normalize event names (case-insensitive)
        available_events = {k.lower(): k for k in epochs.event_id}

        for region, ch_list in regions.items():
            valid_chs = [ch for ch in ch_list if ch in epochs.ch_names]
            if not valid_chs:
                continue

            for cond in conditions:
                cond_lower = cond.lower()
                if cond_lower not in available_events:
                    continue

                evoked = epochs[available_events[cond_lower]].copy().pick_channels(valid_chs).average()
                region_data[cond][region].append(evoked.data.mean(axis=0))

                if times is None:
                    times = evoked.times

    except Exception as e:
        print(f"[{pid}] ⚠️ Error loading: {e}")
        continue

# === Plot: 3×3 Grid with Bootstrapped 95% CI ===
fig, axes = plt.subplots(3, 3, figsize=(14, 9), sharex=True, sharey=True)
axes = axes.flatten()

for i, region in enumerate(region_order):
    ax = axes[i]

    dat_N = np.stack(region_data['Wd2N'][region]) if region_data['Wd2N'][region] else None
    dat_E = np.stack(region_data['Wd2E'][region]) if region_data['Wd2E'][region] else None

    if dat_N is None or dat_E is None:
        ax.set_visible(False)
        continue

    # === Bootstrap confidence intervals ===
    boot_N = bootstrap((dat_N,), np.mean, axis=0, n_resamples=n_resamples, confidence_level=ci_level)
    boot_E = bootstrap((dat_E,), np.mean, axis=0, n_resamples=n_resamples, confidence_level=ci_level)

    mean_N = dat_N.mean(axis=0)
    mean_E = dat_E.mean(axis=0)

    # === Plot with CIs ===
    ax.plot(times, mean_N, color='black', linestyle='-', label='Normal Context')
    ax.fill_between(times, boot_N.confidence_interval.low, boot_N.confidence_interval.high,
                    color='black', alpha=0.25)

    ax.plot(times, mean_E, color='black', linestyle='--', label='Slowed Context')
    ax.fill_between(times, boot_E.confidence_interval.low, boot_E.confidence_interval.high,
                    color='gray', alpha=0.25)

    ax.axvspan(0.100, 0.160, color='skyblue', alpha=0.3)   # N100
    ax.axvspan(0.240, 0.300, color='lightpink', alpha=0.3) # N280
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.set_title(region, fontsize=10)

# Hide unused axes
for j in range(len(region_order), len(axes)):
    axes[j].set_visible(False)

# === Final touches ===
fig.suptitle('Grand Average ERP with 95% CI — sd5 (Wd2E vs Wd2N)', fontsize=15)
fig.text(0.5, 0.04, 'Time (s)', ha='center')
fig.text(0.04, 0.5, 'Amplitude (µV)', va='center', rotation='vertical')
fig.tight_layout(rect=[0.05, 0.05, 1, 0.95])

# ✅ Save plot
save_path = os.path.join(root_dir, 'sd5O1_bootstrap_ERP_3x3_grid.png')
fig.savefig(save_path, dpi=300, bbox_inches='tight')
print(f"\n📊 Saved plot to: {save_path}")

# Show it
plt.show()
```

```python
import os
import mne
import gc
import traceback

# === Paths and constants ===
root_dir = '/content/drive/MyDrive/2025_Summer (Distal Speech Rate)/Project_Code/EEGData'

# Only include participant folders that are numeric and >= 10
participants = sorted([
    p for p in os.listdir(root_dir)
    if os.path.isdir(os.path.join(root_dir, p)) and p.isdigit() and int(p) >= 10
])

# Use CRTE and CRTN as the critical region onset markers
event_id = {'CRTE': 1, 'CRTN': 2}  # Slowed and Normal context

# Epoch time window and baseline
tmin, tmax = -0.1, 0.8  # Epoch from -100ms to +800ms
baseline = (tmin, 0)    # Baseline correction from -100ms to 0ms

def create_and_save_epochs(participant):
    participant_path = os.path.join(root_dir, participant)
    raw_file = os.path.join(participant_path, f"{participant}_O1_ica_reref_raw.fif")
    epo_file = os.path.join(participant_path, f"{participant}_O1_CRITICAL_REGION_epo.fif")

    if not os.path.exists(raw_file):
        print(f"[{participant}] ❌ ICA reref file not found.")
        return

    if os.path.exists(epo_file):
        print(f"[{participant}] ✅ Epochs already exist, skipping.")
        return

    try:
        print(f"[{participant}] Creating epochs...")

        # Load raw EEG data
        raw = mne.io.read_raw_fif(raw_file, preload=True)
        raw.set_montage('GSN-HydroCel-128')

        # Extract events from annotations based on CRTE/CRTN
        events, found_ids = mne.events_from_annotations(raw, event_id=event_id)

        if len(events) == 0:
            print(f"[{participant}] ⚠️ No CRTE/CRTN events found.")
            return

        print(f"[{participant}] Found {len(events)} critical region events.")

        # Create epochs
        epochs = mne.Epochs(
            raw,
            events,
            event_id=event_id,
            tmin=tmin,
            tmax=tmax,
            baseline=baseline,
            reject_by_annotation=True,
            preload=True
        )

        # Save epochs
        epochs.save(epo_file, overwrite=True)
        print(f"[{participant}] ✅ Epochs saved to {epo_file}")

        # Clean up
        del raw, epochs
        gc.collect()

    except Exception as e:
        print(f"[{participant}] ⚠️ Error during epoching: {e}")
        traceback.print_exc()

# === Run epoching for all valid participants ===
for pid in participants:
    create_and_save_epochs(pid)
```

```python
import os
import mne
import matplotlib.pyplot as plt
import gc

# === Root directory ===
root_dir = '/content/drive/MyDrive/2025_Summer (Distal Speech Rate)/Project_Code/EEGData'

# === (1) Optional: Threshold Check on Participant 10 ===
participant = "10"
epo_path = os.path.join(root_dir, participant, f"{participant}_O1_CRITICAL_REGION_epo.fif")

thresholds = [100e-6, 125e-6, 150e-6, 175e-6, 200e-6]
drop_rates = []

if os.path.exists(epo_path):
    print(f"📂 Checking thresholds for: {epo_path}")
    for th in thresholds:
        try:
            epochs = mne.read_epochs(epo_path, preload=True)
            n_total = len(epochs)
            epochs.drop_bad(reject=dict(eeg=th))
            n_clean = len(epochs)
            drop_rate = 1 - (n_clean / n_total)
            drop_rates.append(drop_rate)
            print(f"🔍 {int(th*1e6)} µV — Dropped {drop_rate*100:.1f}% of epochs")
        except Exception as e:
            print(f"⚠️ Error at {int(th*1e6)} µV: {e}")
            drop_rates.append(None)

    # Plot
    plt.figure(figsize=(8, 5))
    plt.plot([int(t*1e6) for t in thresholds], [r * 100 for r in drop_rates], marker='o')
    plt.xlabel('Threshold (µV)')
    plt.ylabel('Dropped Epochs (%)')
    plt.title('Drop Rate vs Threshold — Participant 10')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print(f"❌ Epoch file not found for participant {participant}")

# === (2) Clean All Participants ≥ 10 ===
rejection_threshold = 150e-6  # Set final threshold here

# Get participant directories >=10 only
subject_dirs = sorted([
    d for d in os.listdir(root_dir)
    if d.isdigit() and int(d) >= 10
], key=lambda x: int(x))

for subj in subject_dirs:
    print(f"\n🧠 Cleaning Participant {subj}")
    subj_path = os.path.join(root_dir, subj)

    epo_path = os.path.join(subj_path, f"{subj}_O1_CRITICAL_REGION_epo.fif")
    clean_path = os.path.join(subj_path, f"{subj}_O1_CRITICAL_REGION_epo_clean.fif")

    if not os.path.exists(epo_path):
        print("❌ Original epoch file not found. Skipping.")
        continue

    if os.path.exists(clean_path):
        print("✅ Cleaned file already exists. Skipping.")
        continue

    try:
        epochs = mne.read_epochs(epo_path, preload=True)
        epochs.drop_bad(reject=dict(eeg=rejection_threshold))
        epochs.save(clean_path, overwrite=True)
        print(f"💾 Cleaned epochs saved: {clean_path}")
        del epochs
        gc.collect()

    except Exception as e:
        print(f"⚠️ Error cleaning participant {subj}: {e}")
```

```python
#WDN2 AND WD2D
import os
import mne
import numpy as np
import pandas as pd
from scipy.stats import ttest_rel

# === Parameters ===
root_dir = '/content/drive/MyDrive/2025_Summer (Distal Speech Rate)/Project_Code/EEGData'
participants = [p for p in os.listdir(root_dir) if p.isdigit() and int(p) >= 10]

event_id = {'Wd2N': 1, 'Wd2E': 2}  # Normal vs. Slowed context
tmin, tmax = -0.1, 0.8
baseline = (tmin, 0)
rejection_criteria = dict(eeg=150e-6)  # 150 µV

freqs = np.linspace(4, 30, 20)
n_cycles = freqs / 2.0
bands = {'Theta': (4, 8), 'Alpha': (8, 12), 'Beta': (13, 30)}

all_results = []

for pid in participants:
    raw_file = os.path.join(root_dir, pid, f'{pid}_O1_ica_reref_raw.fif')
    if not os.path.exists(raw_file):
        print(f"[{pid}] ❌ Missing raw file")
        continue

    try:
        raw = mne.io.read_raw_fif(raw_file, preload=True)
        raw.set_montage('GSN-HydroCel-128')

        # Extract events from annotations
        events, _ = mne.events_from_annotations(raw, event_id=event_id)

        # Epoching with rejection
        epochs = mne.Epochs(
            raw, events, event_id=event_id,
            tmin=tmin, tmax=tmax,
            baseline=baseline,
            reject_by_annotation=True,
            reject=rejection_criteria,
            preload=True
        )

        # Optional: add fake 'report' metadata for demo
        if epochs.metadata is None:
            metadata = pd.DataFrame({
                'context': ['Wd2E' if e == 2 else 'Wd2N' for e in epochs.events[:, 2]],
                'report': np.random.choice(['Reported', 'Unreported'], size=len(epochs))
            })
            epochs.metadata = metadata

        # Compute spectral power (not averaged)
        power = mne.time_frequency.tfr_multitaper(
            epochs, freqs=freqs, n_cycles=n_cycles,
            use_fft=True, return_itc=False,
            average=False
        )

        # Crop to time window and average over time
        power.crop(tmin=0, tmax=0.8)
        power_data = power.data.mean(axis=-1)  # shape: (n_epochs, freqs, channels)

        # Extract band power per trial
        for i, trial_meta in epochs.metadata.iterrows():
            for band, (fmin, fmax) in bands.items():
                f_idx = np.where((freqs >= fmin) & (freqs <= fmax))[0]
                value = power_data[i, f_idx, :].mean()  # average freq & channel
                all_results.append({
                    'participant': pid,
                    'context': trial_meta['context'],
                    'report': trial_meta['report'],
                    'band': band,
                    'power': value
                })

    except Exception as e:
        print(f"[{pid}] ⚠️ Error: {e}")
        continue

# === Convert to DataFrame
df_power = pd.DataFrame(all_results)
print(df_power.head())
```

```python
# === Spectral Power Analysis: Wd2N vs. Wd2E ===
import os
import numpy as np
import pandas as pd
import mne
from mne.time_frequency import tfr_multitaper
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt
import seaborn as sns

# === Settings ===
root_dir = '/content/drive/MyDrive/2025_Summer (Distal Speech Rate)/Project_Code/EEGData'
participants = [str(p) for p in range(10, 28) if p != 16]
bands = {'Theta': (4, 8), 'Alpha': (8, 13), 'Beta': (13, 30)}

# === Containers ===
all_data = []

for pid in participants:
    print(f"Processing participant {pid}...")
    participant_path = os.path.join(root_dir, pid)

    try:
        all_fifs = [f for f in os.listdir(participant_path) if f.endswith('.fif')]
    except FileNotFoundError:
        print(f"  ❌ Folder not found for participant {pid}")
        continue

    epo_file = next((f for f in all_fifs if 'epo_clean' in f), None)
    if not epo_file:
        print(f"  ❌ No cleaned epoch file found for {pid}")
        continue

    epo_path = os.path.join(participant_path, epo_file)
    try:
        epochs = mne.read_epochs(epo_path, preload=True)
    except Exception as e:
        print(f"  ⚠️ Error reading {epo_file}: {e}")
        continue

    if epochs.events is None or len(epochs.events) == 0:
        print(f"  ⚠️ No events found for {pid}")
        continue

    # === Add synthetic metadata ===
    epochs.metadata = pd.DataFrame({
        'context': ['Wd2N' if e == 1 else 'Wd2E' for e in epochs.events[:, 2]]
    })

    # === Time–Frequency Decomposition ===
    freqs = np.arange(4, 31, 1)
    n_cycles = freqs / 2.0  # adaptive wavelet length
    power = tfr_multitaper(epochs, freqs=freqs, n_cycles=n_cycles,
                           return_itc=False, average=False)  # shape: (n_epochs, n_channels, n_freqs, n_times)

    # === Band averaging per trial
    for trial_idx in range(len(epochs)):
        context = epochs.metadata.iloc[trial_idx]['context']
        trial_data = power.data[trial_idx]  # shape: (n_channels, n_freqs, n_times)

        for band, (fmin, fmax) in bands.items():
            freq_mask = (power.freqs >= fmin) & (power.freqs < fmax)
            band_data = trial_data[:, freq_mask, :]  # shape: channels × selected_freqs × times
            band_power = band_data.mean()  # mean over all dims
            all_data.append({
                'participant': pid,
                'band': band,
                'context': context,
                'power': band_power
            })

# === Create DataFrame ===
df_power = pd.DataFrame(all_data)

if df_power.empty:
    print("\n🚫 No data available — check participant folders or epoch files.")
else:
    # === Run Paired T-Tests: Wd2E vs Wd2N
    def run_ttests(df, compare_col, levels):
        results = []
        for band in df['band'].unique():
            df_band = df[df['band'] == band]
            df_wide = df_band.pivot_table(index='participant', columns=compare_col, values='power', aggfunc='mean')
            if set(levels).issubset(df_wide.columns):
                tval, pval = ttest_rel(df_wide[levels[0]], df_wide[levels[1]])
                results.append({
                    'Band': band,
                    f"Mean {levels[0]}": df_wide[levels[0]].mean(),
                    f"Mean {levels[1]}": df_wide[levels[1]].mean(),
                    't-value': tval,
                    'p-value': pval
                })
        return pd.DataFrame(results)

    # === T-test: Context Comparison
    context_stats = run_ttests(df_power, 'context', ('Wd2E', 'Wd2N'))

    # === Plotting
    sns.set(style="whitegrid")
    plt.figure(figsize=(10, 5))
    sns.barplot(data=df_power, x='band', y='power', hue='context', ci=68)
    plt.title('Spectral Power by Band and Context Condition (Wd2E vs. Wd2N)')
    plt.tight_layout()
    plt.show()

    # === Display Results
    print("\nT-Test Results: Wd2E vs. Wd2N")
    print(context_stats)

```
