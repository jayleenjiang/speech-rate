```python
import os, mne, numpy as np, gc, traceback

# === CONFIGURATION ===
root_dir = '/content/drive/MyDrive/2025_Summer (Distal Speech Rate)/Project_Code/EEGData'
participants = [f"{i:02}" for i in range(3, 28)]  # Participant IDs: '03' to '27'
event_id = {'Wd2N': 1, 'Wd2E': 2}
tmin, tmax = -0.1, 0.8  # Epoch window
baseline = (tmin, 0)
z_thresh = 3  # Epoch rejection threshold

def create_epochwise_z_cleaned(participant):
    p_dir = os.path.join(root_dir, participant)
    raw_file = os.path.join(p_dir, f"{participant}_O1_ica_reref_raw.fif")
    epo_file = os.path.join(p_dir, f"{participant}_O1_{z_thresh}sd_epo.fif")

    # Skip if already done
    if os.path.exists(epo_file):
        print(f"[{participant}] ⏭️ Already processed: {epo_file}")
        return

    if not os.path.exists(raw_file):
        print(f"[{participant}] ❌ Missing raw file: {raw_file}")
        return

    try:
        print(f"[{participant}] 🚧 Creating cleaned epochs (±{z_thresh} SD)...")

        raw = mne.io.read_raw_fif(raw_file, preload=True)
        raw.set_montage('GSN-HydroCel-128')
        events, _ = mne.events_from_annotations(raw, event_id=event_id)

        epochs = mne.Epochs(
            raw, events, event_id=event_id,
            tmin=tmin, tmax=tmax, baseline=baseline,
            preload=True, reject_by_annotation=True
        )

        data = epochs.get_data()  # (n_epochs, n_channels, n_times)

        # Epoch-wise z-score rejection
        bad_indices = []
        for i, ep in enumerate(data):
            z = (ep - np.mean(ep)) / np.std(ep)
            if np.max(np.abs(z)) > z_thresh:
                bad_indices.append(i)

        print(f"[{participant}] 🧹 Rejected {len(bad_indices)} of {len(data)} epochs")
        if bad_indices:
            print(f"[{participant}] ⛔ Rejected epoch indices: {bad_indices}")
        print(f"[{participant}] ✅ Remaining epochs: {len(data) - len(bad_indices)}")

        # Drop rejected epochs
        epochs.drop(bad_indices)

        # Save cleaned epochs
        epochs.save(epo_file, overwrite=True)
        print(f"[{participant}] 💾 Saved to: {epo_file}")

        del raw, epochs
        gc.collect()

    except Exception as e:
        print(f"[{participant}] ⚠️ Error while processing: {e}")
        traceback.print_exc()

# === EXECUTE FOR ALL PARTICIPANTS ===
for pid in participants:
    create_epochwise_z_cleaned(pid)
```

```python
import os, mne, numpy as np, gc, traceback

# === CONFIGURATION ===
root_dir = '/content/drive/MyDrive/2025_Summer (Distal Speech Rate)/Project_Code/EEGData'
participants = [f"{i:02}" for i in range(3, 28)]  # '03' to '27'
event_id = {'Wd2N': 1, 'Wd2E': 2}
tmin, tmax = -0.1, 0.8
baseline = (tmin, 0)
z_thresh = 4  # ← Per-participant SD threshold

def create_epochwise_z_cleaned(participant):
    p_dir = os.path.join(root_dir, participant)
    raw_file = os.path.join(p_dir, f"{participant}_O1_ica_reref_raw.fif")
    epo_file = os.path.join(p_dir, f"{participant}_O1_{z_thresh}sd_epo.fif")

    if os.path.exists(epo_file):
        print(f"[{participant}] ⏭️ Already processed: {epo_file}")
        return
    if not os.path.exists(raw_file):
        print(f"[{participant}] ❌ Missing raw file: {raw_file}")
        return

    try:
        print(f"[{participant}] 🚧 Processing (±{z_thresh} SD per epoch)...")

        raw = mne.io.read_raw_fif(raw_file, preload=True)
        raw.set_montage('GSN-HydroCel-128')
        events, _ = mne.events_from_annotations(raw, event_id=event_id)

        epochs = mne.Epochs(
            raw, events, event_id=event_id,
            tmin=tmin, tmax=tmax, baseline=baseline,
            preload=True, reject_by_annotation=True
        )

        data = epochs.get_data()  # (n_epochs, n_channels, n_times)

        # Flatten each epoch and compute z-score over all values
        bad_indices = []
        for i, ep in enumerate(data):
            flat = ep.flatten()
            z = (flat - np.mean(flat)) / np.std(flat)
            if np.max(np.abs(z)) > z_thresh:
                bad_indices.append(i)

        print(f"[{participant}] 🧹 Rejected {len(bad_indices)} of {len(data)} epochs")
        if bad_indices:
            print(f"[{participant}] ⛔ Rejected indices: {bad_indices}")
        print(f"[{participant}] ✅ Remaining: {len(data) - len(bad_indices)} epochs")

        epochs.drop(bad_indices)
        epochs.save(epo_file, overwrite=True)
        print(f"[{participant}] 💾 Saved: {epo_file}")

        del raw, epochs
        gc.collect()

    except Exception as e:
        print(f"[{participant}] ⚠️ Error while processing: {e}")
        traceback.print_exc()

# === EXECUTE FOR ALL PARTICIPANTS ===
for pid in participants:
    create_epochwise_z_cleaned(pid)
```
