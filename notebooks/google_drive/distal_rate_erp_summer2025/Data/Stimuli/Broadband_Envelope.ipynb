```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.signal import hilbert, spectrogram
import glob
from scipy.stats import sem  # Standard error of the mean
from scipy.stats import sem

stim_dir = os.path.join(root_dir, 'Data', 'Stimuli')

# 1: Load the .wav file
wav_file_path = os.path.join(stim_dir, 'SR2_N.wav')  # Replace with your actual filename
sample_rate, audio_data = wavfile.read(wav_file_path)

# Convert to mono if stereo
if audio_data.ndim > 1:
    audio_data = np.mean(audio_data, axis=1)

# 2: Compute and plot the spectrogram
f, t, Sxx = spectrogram(audio_data, fs=sample_rate, nperseg=1024, noverlap=992, scaling='spectrum', mode='magnitude')

plt.figure(figsize=(10, 4))
plt.pcolormesh(t, f, Sxx, shading='gouraud') #10 * np.log10(Sxx + 1e-10), shading='gouraud')
plt.colorbar(label='Power')# (dB)')
plt.title('Spectrogram')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [s]')
plt.tight_layout()
plt.show()
```

```python

# 3: Compute the broadband amplitude envelope
print(Sxx.shape)
print(audio_data.shape)
amplitude_envelope = Sxx.mean(0)#np.abs(analytic_signal)

# Plot the envelope
time_vector = np.arange(len(audio_data)) / sample_rate
amp_time = max(time_vector) * np.arange(len(amplitude_envelope)) / len(amplitude_envelope)
plt.figure(figsize=(10, 3))
plt.plot(time_vector, audio_data)
plt.plot(amp_time, 200*amplitude_envelope)
plt.title('Broadband Amplitude Envelope')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()

# 4: Save the envelope to a .txt file
envelope_filename = wav_file_path.replace('.wav', '_env.txt')
np.savetxt(envelope_filename, amplitude_envelope)

print(f"Envelope saved to: {envelope_filename}")
```
