```python
% === USER DEFINED SETTINGS ===
filename = 'SpeechRate003.raw';
n_channels = 107;
sampling_rate = 1000;  % Hz
data_type = 'float32';

% === STEP 1: Load binary EEG data ===
fid = fopen(filename, 'rb');
data = fread(fid, [n_channels, inf], data_type);
fclose(fid);

% === STEP 2: Create EEGLAB EEG structure ===
EEG = pop_importdata('dataformat', 'array', ...
                     'data', data, ...
                     'setname', 'SpeechRate003', ...
                     'srate', sampling_rate, ...
                     'nbchan', n_channels);

% === STEP 3: Add channel labels (optional) ===
for i = 1:n_channels
    EEG.chanlocs(i).labels = sprintf('EEG%d', i);
end

% === STEP 4: Launch EEGLAB GUI and save ===
eeglab;
pop_saveset(EEG, 'filename', 'SpeechRate003.set');
```
