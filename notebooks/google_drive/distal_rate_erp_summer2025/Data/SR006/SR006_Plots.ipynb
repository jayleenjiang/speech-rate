```python
root_dir = '/content/drive/My Drive/Distal_Rate_ERP_Summer2025/'
import os
print(os.listdir(root_dir))
data_dir = root_dir + 'Data/'
```

```python
# prompt: Load .raw file from EEGlab into MNE, and list the contents.

# You might need to install the MNE library first if you haven't
!pip install mne

import mne

# Replace 'your_eeglab_file.set' with the actual path to your .set or .fdt file
# MNE can load various EEGlab formats (.set, .fdt)
eeglab_file_path = data_dir + 'SR003/SpeechRate003.raw'

try:
  # Load the data from the EEGlab file
  raw = mne.io.read_raw_eeglab(eeglab_file_path, preload=True)

  # List the contents of the loaded data
  print(raw)
  print("\nChannels:")
  print(raw.ch_names)
  print("\nInfo:")
  print(raw.info)

except FileNotFoundError:
  print(f"Error: The file '{eeglab_file_path}' was not found.")
except Exception as e:
  print(f"An error occurred while loading the EEGlab file: {e}")
```
