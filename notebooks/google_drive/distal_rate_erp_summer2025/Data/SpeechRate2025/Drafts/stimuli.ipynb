```python
!pip install mne --quiet

import os
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wav
from scipy.signal import hilbert
from scipy.ndimage import gaussian_filter1d
from mne.time_frequency import morlet
import glob
from scipy.signal import convolve
from scipy.io import loadmat
from mne import *
```

```python
root_dir = '/content/drive/My Drive/Distal_Rate_ERP_Summer2025/Data/Stimuli/'
# wav_files = sorted(glob.glob(os.path.join(root_dir, '*.wav'))) # all

# excluded numbers
excluded_numbers = {7, 11, 12, 13, 19, 24, 37, 45, 51, 52, 54, 56, 57, 61, 69, 72, 75, 78}

wav_files = sorted([
    f for f in glob.glob(os.path.join(root_dir, 'SR*_*.wav')) # Match SR*_*.wav
    if f.endswith('_N.wav') or f.endswith('_E.wav')
    and int(os.path.basename(f).split('_')[0][2:]) >= 2 # Extract number after 'SR' and check if >= 2
    and int(os.path.basename(f).split('_')[0][2:]) <= 79 # Extract number after 'SR' and check if <= 79
    and int(os.path.basename(f).split('_')[0][2:]) not in excluded_numbers # Exclude numbers from the set
])
```

```python
smooth_ms = 10

for wav_path in wav_files:
    # load audio
    fs, data = wav.read(wav_path)
    if data.ndim > 1:
        # RMS
        data = np.sqrt(np.mean(data.astype(np.float32)**2, axis=1))
    else:
        data = data.astype(np.float32)

    data = data / np.max(np.abs(data)) # normalize

    # compute amplitude envelope via Hilbert transform
    analytic = hilbert(data)
    envelope = np.abs(analytic)

    # smooth the envelope with Gaussian filter
    sigma = (smooth_ms / 1000) * fs  # convert ms to samples
    envelope_smooth = gaussian_filter1d(envelope, sigma)

    # compute first derivative of envelope
    dt = 1.0 / fs
    envelope_derivative = np.gradient(envelope_smooth, dt)

    # time axis
    t = np.arange(len(data)) / fs

    # plot waveform and smoothed envelope
    plt.figure(figsize=(12, 5))
    plt.plot(t, data, linewidth=0.5, label='Waveform')
    plt.plot(t, envelope_smooth, linewidth=2, label='Envelope')
    plt.plot(t, envelope_derivative/10, color='blue', label='Envelope Derivative')
    plt.title('Broadband Amplitude Envelope — ' + os.path.basename(wav_path))
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.legend(loc='upper right')
    plt.tight_layout()
    plt.show()

    # plot first derivative of envelope
    plt.figure(figsize=(12, 4))
    plt.plot(t, envelope_derivative, color='orange', label='Envelope Derivative')
    plt.xlabel('Time [s]')
    plt.ylabel('dEnvelope/dt')
    plt.title('First Derivative of Amplitude Envelope — ' + os.path.basename(wav_path))
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
from scipy.signal import resample

freqs = np.linspace(100, 15000, 500)  # Adjusted frequency range to include higher frequencies
# freqs = np.array([60])
n_cycles = 7                    # Number of cycles per wavelet
smooth_ms = 10
target_fs = 15000 # Target sampling frequency

for wav_path in wav_files:
    fs, data = wav.read(wav_path)
    if data.ndim > 1:
        data = np.sqrt(np.mean(data.astype(np.float32)**2, axis=1))
    else:
        data = data.astype(np.float32)

    data = data / np.max(np.abs(data)) # normalize

    # Compute wavelets
    sfreq = fs

    # Create Morlet wavelets and apply
    wavelet_output = morlet(sfreq=sfreq, freqs=freqs, n_cycles=n_cycles)

    # Convolve each wavelet with signal and extract amplitude
    power = np.empty((len(freqs), len(data)))
    for i, w in enumerate(wavelet_output):
      conv = convolve(data, w, mode='same')  # shape will match data
      power[i, :] = np.abs(conv)

    # Average amplitude across frequencies
    envelope_wavelet_avg = power.mean(axis=0)

    # Smooth the envelope
    sigma = (smooth_ms / 1000.0) * fs
    envelope_smooth = gaussian_filter1d(envelope_wavelet_avg, sigma)

    # First derivative of smoothed envelope
    dt = 1.0 / fs
    envelope_derivative = np.gradient(envelope_smooth, dt)

    # Resample data to target_fs
    num_samples_target = int(len(data) * (target_fs / fs))
    data_resampled = resample(data, num_samples_target)
    envelope_smooth_resampled = resample(envelope_smooth, num_samples_target)
    envelope_derivative_resampled = resample(envelope_derivative, num_samples_target)

    # Time axis for resampled data
    t_resampled = np.arange(num_samples_target) / target_fs

    # plot waveform and smoothed envelope
    plt.figure(figsize=(12, 5))
    plt.plot(t_resampled, data_resampled, linewidth=0.5, label='Resampled Waveform')
    plt.plot(t_resampled, envelope_smooth_resampled, linewidth=2, label='Resampled Wavelet Envelope')
    plt.title('Resampled Wavelet-based Envelope — ' + os.path.basename(wav_path))
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.tight_layout()
    plt.show()

    # plot first derivative of envelope
    plt.figure(figsize=(12, 4))
    plt.plot(t_resampled, envelope_derivative_resampled, color='orange', label='Resampled Envelope Derivative')
    plt.xlabel('Time [s]')
    plt.ylabel('dEnvelope/dt')
    plt.title('Resampled First Derivative of Wavelet Envelope — ' + os.path.basename(wav_path))
    plt.legend()
    plt.tight_layout()
    plt.show()
```
