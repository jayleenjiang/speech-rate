```python
root_dir = '/content/drive/MyDrive/Distal_Rate_ERP_Summer2025/Data/SpeechRate2025/EEGData'

# Function to load and preprocess one raw EEG file
def load(path):
    raw = mne.io.read_raw_eeglab(path, preload=True)

    return raw

# Function to compare two sessions
def compare_sessions(participant_id):
    participant_path = os.path.join(root_dir, participant_id)
    set_files = sorted([f for f in os.listdir(participant_path) if f.endswith('.set')])

    print(f"\n--- Comparing sessions for participant {participant_id} ---")
    raw1 = load(os.path.join(participant_path, set_files[0]))
    raw2 = load(os.path.join(participant_path, set_files[2]))

    # Make sure data are aligned
    if raw1._data.shape != raw2._data.shape:
        print("❌ Data shapes differ.")
        print(f"Session 1 shape: {raw1._data.shape}, Session 2 shape: {raw2._data.shape}")
        return

    # Subtract the data arrays
    diff = raw1._data - raw2._data
    max_abs_diff = np.max(np.abs(diff))
    mean_abs_diff = np.mean(np.abs(diff))

    # Check if difference is negligible
    if max_abs_diff < 1e-10:
        print(" ⚠️ Sessions are identical.")
    else:
        print(f"Sessions differ. Max abs diff: {max_abs_diff:.2e}, Mean abs diff: {mean_abs_diff:.2e}")
```

```python
compare_sessions('10')
```
